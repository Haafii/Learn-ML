Pandas Data Frame:
Pandas DataFrame is two-dimensional tabular data structure with labeled axes (rows and columns).

```python
# import library 
import pandas as pd
import numpy as np
```

```python
# importing California Housing Dataset
from sklearn.datasets import fetch_california_housing

california_housing = fetch_california_housing()
print(type(california_housing))
```

```python
california_housing
```

```python
# pandas DataFrame
house_df = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)
house_df.head() 
```

```python
house_df.shape # (20640 rows, 8 columns)
```

```python
# importing data from CSV file to pandas dataFrame
diabetes_df = pd.read_csv('diabetes.csv')
diabetes_df.head()
```
