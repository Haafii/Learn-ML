```python
# importing libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# setting a theme for the plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips, x ='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_length',y='petal_width',hue='species',data=iris)
```

```python
# loading the titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# count plot
sns.countplot(x='class',data=titanic)
```

```python
sns.countplot(x='survived', data=titanic)
```

```python
# bar plot
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
# House price dataset
from sklearn.datasets import fetch_california_housing
california_housing = fetch_california_housing()
house_df = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)
house_df['PRICE'] = california_housing.target
house_df.head()
```

```python
# Distribution plot
sns.displot(house_df['PRICE'],kde=True)
```

```python
# correlation using heat map
corr = house_df.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(corr  , cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
plt.show()
```
